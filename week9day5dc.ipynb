```python
import seaborn as sns
iris = sns.load_dataset('iris')
```

```python
import matplotlib.pyplot as plt
plt.hist(iris['petal_length'])
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Petal Lengths')
plt.show()
```

```python
sns.pairplot(iris)
```
